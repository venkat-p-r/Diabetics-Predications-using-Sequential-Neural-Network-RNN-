```python
import numpy as np
from numpy import loadtxt
```

```python
data= loadtxt('diabetesclean.csv',delimiter=',') # to read csv file
```

```python
data.shape
```

```python
x= data[:,0:8] # Data
y= data[:,8]   # Label
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model= Sequential()
```

```python
model.add(Dense(8,input_dim=8 , activation ='relu')) # input 8 neurons shape relu is weight of the neurons
model.add(Dense(8,activation = 'relu')) # Hidden neuron layer
model.add(Dense(1,activation ='sigmoid')) # Output layer ,the sigmoid is used for only 2 output like binary classification.
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=500,batch_size=20) # epoches of number iteration of train data
```

```python
_,acc=model.evaluate(xtest,ytest)
print('Accuracy score of Test Data:',acc*100)
```
